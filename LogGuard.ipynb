```python
# Change 'Level' and 'Component' to categorical
df['Level'] = df['Level'].astype('category')
df['Component'] = df['Component'].astype('category')
```

```python
# List of words to detect in the Content field
incident_keywords = ['Error', 'Failed', 'Invalid']

df['Incident'] = df['Content'].apply(lambda c: any(keyword in c for keyword in incident_keywords)).astype(int)
```

```python
from scipy.sparse import hstack, csr_matrix
from sklearn.feature_extraction.text import TfidfVectorizer


X_other_features = df[['Level', 'Component']]
text_column = df['Content']
```

```python
# One-hot encode the categorical features
encoder = OneHotEncoder()
X_other_features_encoded = encoder.fit_transform(X_other_features)
```

```python
from scipy.sparse import hstack, csr_matrix
from sklearn.feature_extraction.text import TfidfVectorizer


# Vectorize the text data
vectorizer = TfidfVectorizer()
X_text_vectorized = vectorizer.fit_transform(text_column)
```

```python
# Combine the vectorized text data with other features
X_combined = hstack([X_text_vectorized, X_other_features_encoded])

# Assuming 'Incident' is your target variable
y = df['Incident']
```

```python
X_combined
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)
```

```python
models = {
    "LR": LogisticRegression(),
    "DT": DecisionTreeClassifier(),
    "DT(max_depth=3)": DecisionTreeClassifier(max_depth=3),
    "KNN": KNeighborsClassifier(),
    "SVC": SVC(),
    "RF": RandomForestClassifier(),
}
```

```python


def train_models():
    best = {'name': None, 'accuracy': 0}
    
    for name, model in models.items():
        # Perform cross-validation
        scores = cross_val_score(model, X_train, y_train, scoring='accuracy', cv=10)
        
        # Calculate mean accuracy
        mean_accuracy = scores.mean()
        
        # Calculate other metrics
        precision = cross_val_score(model, X_train, y_train, scoring='precision', cv=10).mean()
        recall = cross_val_score(model, X_train, y_train, scoring='recall', cv=10).mean()
        f1 = cross_val_score(model, X_train, y_train, scoring='f1', cv=10).mean()
        roc_auc = cross_val_score(model, X_train, y_train, scoring='roc_auc', cv=10).mean()

        print(f"\nModel {name}")
        print(f"\tAccuracy: {mean_accuracy:.2%}")
        print(f"\tPrecision: {precision:.2f}")
        print(f"\tRecall: {recall:.2f}")
        print(f"\tF1 Score: {f1:.2f}")
        print(f"\tROC AUC: {roc_auc:.2f}")

        if mean_accuracy > best['accuracy']:
            best = {'name': name, 'accuracy': mean_accuracy}

    print(f"\nBest Performing Model {best['name']} with (mean CV of) accuracy = {best['accuracy']:.2%}")
```

```python
train_models()
```

```python
# Create and train the model
modelRF = RandomForestClassifier()
modelRF.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = modelRF.predict(X_test)
```

```python
# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred, zero_division=1))
```
